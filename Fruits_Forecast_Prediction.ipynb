```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# k>1
# k!=even
# k=sqrt(25)=5
# Hence,we will decide on 5 nearest neighbors
```

```python
#Linear Regrresion Classifiers
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,metric='euclidean')
```

```python
# Different Classifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# KNeighborsClassifier
step1=ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,3,4])
    ],remainder='passthrough')
step2=KNeighborsClassifier()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(x,y)
y_pred=pipe.predict(x)
y_pred
print("Classification report",classification_report(y,y_pred))
print("Accuracy score",accuracy_score(y,y_pred))
print("Confusion matrix",confusion_matrix(y,y_pred))
```

```python
# DecisionTreeClassifier
step1=ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,3,4])
    ],remainder='passthrough')
step2=DecisionTreeClassifier()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(x,y)
y_pred=pipe.predict(x)
y_pred
print("Classification report",classification_report(y,y_pred))
print("Accuracy score",accuracy_score(y,y_pred))
print("Confusion matrix",confusion_matrix(y,y_pred))
```

```python
# GradientBoostingClassifier
step1=ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,3,4])
    ],remainder='passthrough')
step2=GradientBoostingClassifier()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(x,y)
y_pred=pipe.predict(x)
y_pred
print("Classification report",classification_report(y,y_pred))
print("Accuracy score",accuracy_score(y,y_pred))
print("Confusion matrix",confusion_matrix(y,y_pred))
```
